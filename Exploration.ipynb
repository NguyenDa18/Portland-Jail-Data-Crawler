```python
import pandas as pd
import json
```

```python
CSV_URL = "csvs/inmate_details.csv"

df = pd.read_csv(CSV_URL)

# sources
# https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173
```

```python
df = df.dropna()
```

```python
df["Charges"] = df["Charges"].apply(eval)
```

```python
df
```

```python
charges = pd.DataFrame(df['Charges'])
```

```python
crime_counts = [len(charges.iloc[i]['Charges']) for i in range(len(df))]
df["Num Charges"] = crime_counts
```

```python
df.sort_values(by='Num Charges', ascending = False)
```
